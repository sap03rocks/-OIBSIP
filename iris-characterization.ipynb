 ***in order to prevent our machine learning model from beign overfitting and just have 100 percent accuracy with no other data getting fitted we split the dataset into two parts the one which is training dataset which gives us the training data for our model and the testing data to predict the accuracy of the model and thus improving chances of error detectionm and so we use scikit library's test train split function***

```python

from sklearn.model_selection import train_test_split
x = df.drop(columns = ['Species'])
y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.80)
y_test
```

```python
#Importing library for decision classifier
from sklearn.tree import DecisionTreeClassifier
id3=DecisionTreeClassifier(criterion='entropy')
#Fit the data
k=id3.fit(x_train,y_train)
#predict the data
y_pred=id3.predict(x_test)
print(y_pred)
```

```python
#Confusion matrix 
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_pred,y_test)
plt.figure(figsize=(10,8))
sb.heatmap(cm,annot=True)
plt.xlabel('predicted-y')
plt.ylabel('actual-y')
plt.show()
```

```python
#Accuracy score and model score
print(classification_report(y_pred,y_test))
print('accuracy-score',accuracy_score(y_pred,y_test))
print('Model score',id3.score(x_test,y_test))
```
