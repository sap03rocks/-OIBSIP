```python
#READING DATASET WITH UNEMPLOYMENT RATE UPTO 11 2020 WHICH IS A CSV FILE USING PANDAS
df2=pd.read_csv('/kaggle/input/unemployment-in-india/Unemployment_Rate_upto_11_2020.csv')
df2
```

```python
sb.pairplot(df)#helps to compare the whole dataset
```

```python
fig = plt.figure(figsize = (10, 10))
sb.histplot(x=' Estimated Unemployment Rate (%)', data=df, kde=True, hue='Area')
plt.title('Unemployment according to Area')
plt.xlabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (8, 6))
sb.lineplot(y=' Estimated Unemployment Rate (%)', x=' Date', data=df)
plt.title('Unemployment according to Date')
plt.xlabel('Date')
plt.xticks(rotation=90)
plt.ylabel('Unemployment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (30, 15))
sb.histplot(x=' Estimated Labour Participation Rate (%)', data=df, kde=True, hue='Area')
plt.title('Labour Participation according to Area')
plt.xlabel('Labour Participation Rate')
plt.show()
```

```python
y=df[' Estimated Unemployment Rate (%)']
x=df['Region']
plt_1 = plt.figure()
plt.title('Umemployment Rate')
plt.xlabel("States")
plt.ylabel("Estimated Unemployment rate",)
plt.xticks(rotation='vertical',fontsize=12)
sb.histplot(x,)
```

```python
fig = plt.figure()
plt.scatter(df2['Region.1'], df2[' Estimated Labour Participation Rate (%)'])
plt.title('Labour Participation according to Region')
plt.xlabel('Region')
plt.ylabel('Labour Participation Rate')
plt.show()
```

```python
sb.histplot(x=' Estimated Employed', data=df, kde=True, hue='Area')
plt.title('Employment according to Area')
plt.xlabel('Employment Rate')
plt.show()
```

```python
fig = plt.figure(figsize = (9, 9))
plt.scatter(df2['Region.1'], df2[' Estimated Employed'])
plt.title('Employment according to Region')
plt.xlabel('Region')
plt.ylabel('Employment Rate')
plt.show()
```
